```python
import matplotlib.pyplot as plt
import numpy as np

# Data for radar chart
categories = ['Question', 'Self-disclosure', 'Affirmation and Reassurance', 'Providing Suggestions', 
              'Others', 'Reflection of feelings', 'Information', 'Restatement or Paraphrasing']

# Normalized data for each category
openai_data = [2.9669720012949563, 3.6102033674254232, 2.446345865029362, 1.7535055000723885, 
               3.476629418777302, 3.1850499259789933, 2.921304631660524, 3.447192081987688]

llama2_data = [2.7540624551182993, 2.2247471349595664, 2.1303019588614127, 1.2036647819236286, 
               2.5972592647004418, 2.243753036638591, 2.3089694000968444, 3.2958215555594372]

gpt4_data = [2.633511681848235, 2.574731686364838, 2.1344329442451486, 1.2900509933554938, 
             2.6911113957934223, 2.5793598769634896, 2.590384937926466, 3.4181230023315514]

mixtral_data = [1.909661045846986, 2.572313038567049, 2.165810891575688, 1.0, 
                1.970826290413385, 2.2619562074683013, 3.2507693031242986, 4.069796982712946]

lora_data = [3.943732807690573, 4.584118603321191, 3.9298772782553035, 2.795386830136, 
             4.010562764708446, 4.936892092622977, 4.2081071595414645, 5.60715027960626]

# Number of variables
num_vars = len(categories)

# plt.style.use('seaborn')
plt.figure(figsize=(10, 8))
# Compute angle for each category
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # complete the loop

# Radar chart
fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(polar=True))

# Draw one axe per variable and add labels
plt.xticks(angles[:-1], categories, color='grey', size=12)

# Draw ylabels
ax.set_rlabel_position(30)
plt.yticks([1, 2, 3, 4, 5, 6], ["1", "2", "3", "4", "5", "6"], color="grey", size=12)
plt.ylim(0, 6)

# Plot data and fill area for each dataset
openai = openai_data + openai_data[:1]
ax.plot(angles, openai, linewidth=1, linestyle='solid', label='gpt-3.5-turbo-1106')
ax.fill(angles, openai, alpha=0.1)

llama2 = llama2_data + llama2_data[:1]
ax.plot(angles, llama2, linewidth=1, linestyle='solid', label='Llama2')
ax.fill(angles, llama2, alpha=0.1)

gpt4 = gpt4_data + gpt4_data[:1]
ax.plot(angles, gpt4, linewidth=1, linestyle='solid', label='gpt-4-1106-preview')
ax.fill(angles, gpt4, alpha=0.1)

mixtral = mixtral_data + mixtral_data[:1]
ax.plot(angles, mixtral, linewidth=1, linestyle='solid', label='Mixtral-8x7b-Instruct')
ax.fill(angles, mixtral, alpha=0.1)

lora = lora_data + lora_data[:1]
ax.plot(angles, lora, linewidth=1, linestyle='solid', label='LoRA-Mistral-7b')
ax.fill(angles, lora, alpha=0.1)

# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show the plot
plt.savefig('radar_chart.png')
```
